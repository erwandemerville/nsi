# Rappels : appartenance √† un m√™me r√©seau

Deux machines appartiennent au m√™me r√©seau si elles poss√®dent la m√™me *adresse r√©seau*. Dans ce cas, elles peuvent communiquer directement.

On rappelle que pour d√©terminer l'adresse r√©seau d'une machine dont on conna√Æt l'adresse IP, il suffit d'appliquer l'op√©rateur logique ET (AND en anglais) bit √† bit entre l'adresse IP et le masque de sous-r√©seau. La table de v√©rit√© de cette op√©rateur est la suivante :

| x | y | x ET y |
| --- | --- | --- |
| 0 | 0 | 0 |
| 1 | 0 | 0 |
| 0 | 1 | 0 |
| 1 | 1 | 1 |


**Exemple** : D√©terminons l'adresse r√©seau de la machine d'adresse IP `193.55.221.62/24`. Le masque est `/24` donc vaut `255.255.255.0` en notation d√©cimale. On peut alors appliquer le ET logique bit √† bit :

|                               | Notation binaire                        | Notation d√©cimale |
|-------------------------------|:--------------------------------------|:--------------------|
| Adresse IP (d'une machine)    | `11000001.00110111.11011101.00111110` | `193.55.221.62`     |
| Masque                        | `11111111.11111111.11111111.00000000` | `255.255.255.0`     |
| Adresse r√©seau (ET bit √† bit) | `11000001.00110111.11011101.00000000` | `193.55.221.0`      |

L'adresse r√©seau de la machine `193.55.221.62/24` est donc `193.55.221.0`.
